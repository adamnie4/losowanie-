```python
import pandas as pd 
import numpy as np 
import random 
```

```python
# plik - podaj nazwe pliku
# ile - ile wierszy ma wylosowac

def losowanie(plik,ile):
    df = pd.read_csv(plik)
    df_row = df.shape[0]
    #Pusta baza danych o takiej samej strukturze jak baza wejsciowa:
    df_empty = df[0:0]
    list_index = []

    while True:
        x = random.randint(0,df_row-1)
    
        if x in list_index:
            pass
        else:
            list_index.append(x)
        
        if len(list_index) == ile+1:
            break
        #Nowa baza danych z wylosowaną próbą: 
        df2 = df.iloc[list_index,]
        
    return df2



    
```
